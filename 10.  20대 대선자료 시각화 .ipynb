```r
pievec <- aggregate(cbind(선거인수 , 투표수) ~ 광역시도, election,sum )
pievec
```

```r
vote <- arrange(pievec,desc(투표수))[,3]
vote
```

```r
region <- as.character(arrange(pievec,desc(투표수))[,1])
region
```

```r
ratio <- round(vote/sum(vote) *100,2)
```

```r
par(family = "AppleGothic")
options(warn = -1)
pie3D(vote,
      main = "광역시도별 투표율",
      labels = paste(region,"\n",ratio,"%"),
      labelcex = 0.75,
      radius = 0.1,
      explode = 1
     )
```

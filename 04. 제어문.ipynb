```r
x <- 10
y <- runif(1) * 100

if(x > y ){
    print("x>runif()")
}
print("--- End ---")
```

```r
x <- 10
y <- runif(1) * 100

if(x > y ){
    print("x>runif()")
}else{
    print("x <= runif()")
}
    print("--- End ---")
```

```r
x <- 10
y <- runif(1) * 100

if(x > y ){
    print("x>runif()")
}else if(x <y ){
    print("x< runif()")
}else{
    print("x = runif()")
}
    print("--- End ---")
```

```r
# 홀수 짝수 구분

a <- c(8,7,6)

if( a %% 2 == 0){
#    cat(a, "는 짝수입니다." , sep= "" )
    sprintf("%i는 짝수 입니다.",a)
}else{
    cat(a, "는 홀수입니다." )
}
```

```r
a <- c(8,7,6)
ifelse(a %% 2 ==0, paste(a,": 짝수"), paste(a,": 홀수"))
```

```r
# 사용자로부터 data를 입력 받아 입력한 숫자가 홀수 인지 짝수 인지 판단하여 출력하기.

num1 <-as.integer(readline("숫자를 입력하세요 : "))

#입력하신 숫자는 12이고 짝수입니다. 
ifelse(num1 %% 2 == 0 , sprintf("입력하신 숫자는 %i이고 짝수입니다.",num1),sprintf("입력하신 숫자는 %i이고 홀수입니다.",num1))
```

```r
score <- as.integer(readline("점수를 입력하세요 : "))

if(score >= 90){
    print("A학점 입니다.")
}else if(score >=80){
    print("B학점 입니다.")
}else if(score >= 70){
    print("C학점 입니다.")
}else if(score >= 60){
    print("D학점 입니다.")
}else{
    print("F학점 입니다.")
}
```

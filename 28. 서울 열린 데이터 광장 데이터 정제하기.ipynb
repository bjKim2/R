```r
rawseoul_test <- rawseoul_test[ !(rawseoul_test$자치구별 == "합계"),]
rawseoul_test <- rawseoul_test[ ,-4]

seoulsamp <- c()
for(i in 2011:2019){
    for (j in as.character(unique(rawseoul_test$자치구별))){
        for(k in 1:12){
            seoulsamp <-rbind(seoulsamp, cbind(기간 = i, 월 = k,자치구별 = j))
        }
    }
    
}
seoulsamp<-as.data.frame(seoulsamp)
```

```r
seoulsamp$발생건수 <- 777
seoulsamp$사망자수 <- 777
seoulsamp$부상자수 <- 777
```

```r
x <-t(rawseoul_test[,3:15])
for(k in 1:9){
    for(q in 1:25){
        for( j in 1:12){
            for(i in 1:3){
                samprow = 300*(k-1)+12*(q-1)+j
                seoulsamp[samprow,3+i] <- x[j+1,i+3*(q-1)+75*(k-1)]
            }
        }
    }
}
```

```r
seoulsamp$사망자수 <-gsub("-","0" ,seoulsamp$사망자수)
seoulsamp$발생건수 <-gsub("-","0" ,seoulsamp$발생건수)
seoulsamp$부상자수 <-gsub("-","0" ,seoulsamp$부상자수)
rownames(seoulsamp) <- c()
for(i in 4:6)
    seoulsamp[,i] <- as.numeric(seoulsamp[,i])
for(i in 1:2)
    seoulsamp[,i] <- as.numeric(seoulsamp[,i])
```

```r
for(i in 4:6)
    print(sum(!(seoul[,i] == seoulsamp[,i])))
```

```r
df <- c()
for (i in 1:(nrow(rawseoul_test)/3) ){

    기간 <- as.numeric(rawseoul_test[i*3,1])
    월 <- c(1:12)
    자치구별 <- as.character(rawseoul_test[i*3,2])
    발생건수 <- as.numeric(t(rawseoul_test[i*3-2,4:15]))
    사망자수 <- t(rawseoul_test[i*3-1,4:15])
    
    사망자수 <- as.numeric(ifelse(사망자수 == '-',0,사망자수))
    부상자수 <- as.numeric(t(rawseoul_test[i*3,4:15]))
    덩어리 <- data.frame(기간,월,자치구별,발생건수,사망자수,부상자수)

    colnames(덩어리)<- c("기간","월","자치구별","발생건수","사망자수","부상자수")
    df<- rbind(df,덩어리)
}
df
```

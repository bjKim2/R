```r
samp <- sample(1:nrow(iris),nrow(iris)*0.7)
```

```r
irisSamp <- iris[samp,]
head(irisSamp)
```

```r
summary(irisSamp) # 스페이시스가 비율이 1:1:1이 아니다. 샘플링 제대로 안됨
```

```r
table(irisSamp$Species)
```

```r
#iris Sampling 시 Species 별로 Sampling

samp <- c(sample(1:50, 35),sample(51:100,35),sample(101:150,35))

```

```r
irisSamp <- iris[samp,]
table(irisSamp$Species)
```

```r
# Training Data와 Test Data 분리하기

iris.train <- iris[samp,]
iris.test <- iris[-samp,]
```

```r
summary(iris.test)
```

#부분집합

```r
#iris 에서 Species 가 setosa 인 것만 추출하여 iris_setosa라는 데이터프레임 만들기

iris_setosa <- subset(iris, Species == "setosa")
```

```r
print(tail(iris_setosa))
iris_setosa$Species <- factor(iris_setosa$Species)
str(iris_setosa)
```

```r
str(iris_setosa)
```

```r
# iris에서 Species가 versicolor인것만 추출하여 iris_versicolor라는 데이터프레임 만들기

iris_versicolor <- subset(iris,Species == "versicolor")
```

```r
str(iris_versicolor)
iris_versicolor$Species <- factor(iris_versicolor$Species)
```

```r
rownames(iris_versicolor) = c()
class(iris_versicolor)
```

```r
str(iris_versicolor)
summary(iris_versicolor)
```

```r
# 정제 포인트는 rownames,str에서 factor 보기
iris_versicolor
```

```r
#iris에서 Species가 setosa이고 Sepal.Length가 5보다 큰 Sample의 갯수는?
sum(iris$Sepal.Length > 5 & iris$Species == "setosa")
nrow(subset(iris, Species == "setosa" & Sepal.Length > 5))
```

```r
# subset으로 컬럼 지우기
subset(iris, select = -Species)
```

# 그룹 연산 함수 

```r
# 틸트 연산자 ~ y=ax +b
# iris에서 Species 기준으로 Sepal.Width의 평균 구하기

aggregate(Sepal.Width ~ Species,iris,mean)
```

#서울시 교통사고 조사 데이터셋

```r
# 데이터셋을 불러와서 seoul 이라는 변수에 넣기

seoul <- read.csv("../Data/newSeoul2019.csv")
seoul
```

```r
tail(seoul)
```

```r
summary(seoul)
```

```r
seoul
arrange(seoul,seoul$년도,seoul$자치구명)
```

```r
sum(is.na(seoul))
for ( i in 1:5){
    cat(colnames(seoul)[i],"컬럼의 NA 인덱스 : " ,grep("@",ifelse(is.na(seoul[,i]),"@","")),"\n")
}
```

```r
summary(seoul)
```

```r
#년, 월, 자치구별 데이터 빈도수 확인
for(i in 1:3)
    print(table(seoul[,i]))
```

```r
# 자치구별 발생건수의 평균

regionAcc.mean <- aggregate(발생건수 ~ 자치구명, seoul, mean)
head(regionAcc.mean)
```

```r
# regionAcc.mean를 발생건수 기준으로 내림차순 정렬하기
arrange(regionAcc.mean, desc(발생건수))
```

```r
# A 컬럼의 최대값, 최소값, 빈도수 구하기

max(test$A)
min(test$A)
table(test$A)
```

```r
# 각 컬럼의 빈도수 구하기
for (i in 1:ncol(test)){
    cat(colnames(test)[i], "컬럼의 빈도 분석 결과  ")
    print(table(test[,i]))
    cat("-----------------------\n")
}
# cat 은 테이블네임즈가 출력이 안되서 print 로 table만 출력함
# 옆으로 체크하는 거는 포문 많이쓰고 밑으로 체크하는건 주로 함수로 쌉가능 
```

```r
# 각 컬럼의 빈도수의 최대값, 최소값 구하기
for (i in 1:ncol(test)){
    cat(colnames(test)[i], "컬럼의 빈도수의 최대값과 최소값\n")
    vec1 <- (table(test[,i]))
    cat("최대값 : ",max(vec1) ,"최소값 : ",min(vec1),"\n-----------------------\n")
}
```

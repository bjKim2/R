```r
### 서울시 교통사고 데이터로 발생건수와 부상자수의 산포도 구하기
seoul <- read.csv("../Data/newSeoul2019.csv")
head(seoul)
```

```r
# 발생건수의 숫자가 150이상이면 빨간색, 아니면 파란색으로 표현하기 
par(family = "AppleGothic")
options(warn = -1)

plot(seoul$발생건수, seoul$부상자수,
     main = "서울시 교통사고 발생건수와 부상자수 산포도",
     xlab = "발생건수" , 
     ylab = "부상자수" ,
     col = ifelse(seoul$발생건수 >= 150, "red", "blue"),
     type = "p"
    )
```

```r
# main title : Simple Time Series
y <- c(1,3,2,4,8,5,7,6,9)
plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series"
    )
     
```

```r
# 꺽은선 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "l" # line
    )
```

```r
# 점이 찍힌 꺽은선 그래프
y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "b" # both
    )
```

```r
#점이 빠진 꺽은선 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "c" # both
    )
```

```r
#점과 꺽은선을 중첩한 그래프
y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "o" # both
    )
```

```r
#점과 꺽은선을 중첩한 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "o", # overlap
     col = c(rep("red",3), "blue",rep("green",6)) # both
    )
```

```r
# 수직선 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "h", # 수직선 그래프
     col = c(rep("red",3), "blue",rep("green",6)) # both
    )
```

```r
# 왼쪽 값을 기초로한 계단 모양 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "s" # step 계단
    )
```

```r
# 오른쪽 값을 기초로한 계단 모양 그래프
y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "S" # step 계단
    )
```

```r
# 오른쪽 값을 기초로한 계단 모양 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "s", # p 는 포인터
     lty = 4 # 1 실선 2 점선 3 촘촘점선 1~6 번까지 있고 그이상은 x%6 으로 함 
    )
```

```r
# 오른쪽 값을 기초로한 계단 모양 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "p", # p 는 포인터
     pch = c(24,20,9) ,# 1 실선 2 점선 3 촘촘점선 1~6 번까지 있고 그이상은 x%6 으로 함 
     col = c("red","blue","yellow")
    )
```

```r
# 오른쪽 값을 기초로한 계단 모양 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "s", # p 는 포인터
     lty = 4 # 1 실선 2 점선 3 촘촘점선 1~6 번까지 있고 그이상은 x%6 으로 함 
    )
```

```r
# 오른쪽 값을 기초로한 계단 모양 그래프

y <- c(1,3,2,4,8,5,7,6,9)

plot(x = 1:length(y),y,
     xlab = "x",
     xlim= c(0,length(y)+1),
     ylim= c(min(y)-1,max(y)+1),
     main = "Simple Time Series",
     type = "p", # p 는 포인터
     pch = c(24,20,9) ,# 1 실선 2 점선 3 촘촘점선 1~6 번까지 있고 그이상은 x%6 으로 함 
     col = c("red","blue","yellow")
    )
```

# 선그래프

```r
# 3종류의 dataset의 선 그래프
abc <- c(260, 300, 250, 280, 310)
def <- c(180, 200, 210, 190, 170)
ghi <- c(210, 250, 260, 210, 270)
```

```r
par(family = 'AppleGothic')
options(warn = -1)

plot(abc,
     type = "o",
     col = "red",
     ylim = c(0,400),
     axes = F, # x,y 좌표 Frame 제거
     ann = F, # x,y 라벨 제거 
    )

# x축과 x축의 라벨을 설정
axis(1, at= 1:5, lab = c("서울",'대전','대구','광주','원주'))

# y축과 y축의 라벨을 설정
axis(2, ylim = c(0,400))

# main title 지정
title(main = "과일 판매량", col.main = "red")

# x축 title(label) 지정
title(xlab = "지역", col.lab = "blue")

# y축 title(label) 지정
title(ylab = "판매량", col.lab = "blue")

# 선그리기 함수로 def 그리기
lines(def,
      type= "o",
      pch = 22,
      col = "green",
      lty = 3
     )

# 선그리기 함수로 ghi 그리기
lines(ghi,
      type = "o",
      pch = 22,
      col = 'blue',
      lty = 2
     )

# 범례 
legend(4,400, # x,y 좌표설정 범례의 왼쪽위 좌표 
       c("야구장", "축구장", "해변가"),
       col = c("red","green","blue"),
       pch = c(21,22,22), # plot character
       lty = c(1,3,2), # line type
       cex = 0.8 # 1이 100% 일 떄 비율값 적음
      )

```

```r
# 막대 그래프는 matrix 만 가능 dataframe 은 불가능, 그래서 as.matrix 를 써야한다.
x <- 1:6 
x
barplot(x, # 현재 x는 벡터
        main = "1~6",
        xlab = "sales",
        ylab = "Amounts",
        names = c("A","B","C","D","E","F"), # 벡터라서 names 함수사용  
        col = c("red",rep("blue",5))
       )
```

```r
# 복잡한 형태의 막대 그래프
# 3행 2열의 matrix
x <- matrix(1:6, 3, 2) # 1:6을 3행 2열로 만들기
x
barplot(x,#누적 막대그래프가 디폴트다. 열이 x축
        ylim = c(0,16)
       ) 
```

```r
# 누적 막대 그래프 분리후 열별 label 붙이기
par(family = "AppleGothic")
options(warn = -1)

rownames(x) <- c("1월","2월","3월")
barplot(x,
        beside = T, # 누적부분 해체
        ylim = c(0,max(x)+2),
        names = c("Korea", "America"), # names 는 2 개 또는 6개 를 줄수 있다
#        names = c("k1","k2","k3","A1","A2","A3") 
        main = "수출량",
        xlab = "국가",
        ylab = "빈도수",
        col = c("red","green","blue"),
        legend.text = T # 막대그래프는 기본적으로 레전드를 가지고있어서 T 만 쓰면 됨
       )
```

```r
par(family = "AppleGothic")
sales <- c(210,110,400,550,600,130)
week <- c("월","화","수","목","금","토")
ratio <- sales / sum(sales) * 100
ratio
```

```r
par(family = "AppleGothic")
pie(sales,
    main = "영업실적 파이차트",
    labels = paste(week,"\n", ratio,"%"),
    col = rainbow(length(sales))
   )
```

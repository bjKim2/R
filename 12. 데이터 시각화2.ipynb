### 한 화면에 차트 여러개 그려서 비교하기

```r
price <- runif(10,1,100)
plot(price
     
    )
```

```r
# 2행 2열 차트 그리기 : type 속성으로 구분해 보기 (같은 셀 안에 있어야함)

par(mfrow = c(2,2))
plot(price, type= "l")
plot(price, type= "o")
plot(price, type= "h")
plot(price, type= "s")


```

```r
# 2행 2열 차트 그리기 : pch 속성으로 구분해 보기

par(mfrow = c(2,2))


plot(price, type= "o", pch=5)
plot(price, type= "o", pch=15)
plot(price, type= "o", pch=20, col = "blue")
plot(price, type= "o", pch=rep(c(15,8),20)
                     , col = "orange"
                     , cex = c(rep(1,3), 3, 1)
                       
    )
```

```r
options(warn = -1)
par(family = "AppleGothic")

# 기타 그래프 
x <- c( 1, 3, 6, 8, 9)
y <- c(12, 56, 78, 32, 9)

plot(x,y)

# 선분 그리기 (3,3) 와 (4,4)
segments(6,78, 8,32, col ="red")

# 화살표 그리기
arrows(3, 56, 1, 12)

# 사각형 그리기
rect(4,20 ,6,30, col = "red", border="red",density = 10)

# 글자 쓰기
text(4,40,"this is a 샘플",cex = 1,srt = 90)

# 테두리 활용하기
mtext("character", side = 2, adj = 1) # adj = 0~1

# chart의 테두리 그리기
box(lty =1, col="red")

# x축의 추가 : y 축의 40위치에 0~10까지,표시하고 색은 빨간색
axis(1, pos = 81,at = 0:10, col = "red")

# y축의 추가 : x 축의 6위치에 20~60까지 표시하고 색은 파란색
axis(2, pos = 6, at= 20:60 , col = "blue")
```

```r
vote_matplot <- t(as.matrix(head(arrange(election,desc(election$투표수)),5)[,5:7]))
```

```r
colnames(vote_matplot) <- paste(head(arrange(election,desc(election$투표수)),5)[,2], head(arrange(election,desc(election$투표수)),5)[,3])
```

```r
vote_matplot
```

```r
par(family = "AppleGothic")
options(scipen = 100)

barplot(vote_matplot,
        main = "투표수 높은 지역의 후보별 투표수",
        col = c("skyblue","red","lightGreen"),
        ylim = c(0,500000),
        legend = T,
        cex.names = 0.7
       )

for( i in 1:5){
    text(i*1.2-0.5,sum(vote_matplot[,i]) + 10000,sum(vote_matplot[,i]) ,cex = 1)
}
```

# 수학 함수의 시각화 
y = x^2

```r
f <- function(x){
    y = x^2
    return(y)
}
```

```r
x <- 1:3
f(x)
```

```r
x <- seq(-5,5,0.1)

plot(x,f(x),
    type="l"
    )
```

y = e^(-x) - 4x^2 + x + 3|x| -3

```r
f <- function(x){
    y = exp(-x) - 4* x^2 + x + 3* abs(x) -3
    return(y)
}
```

```r
plot(x,f(x),
     type = "l"
    
    )
```

```r
# y = x^2 + 2x + 3
f <- function(x){
    y = x^2 + 2*x + 3
}
plot(x,f(x),
    type = "l"
    )

lines(x, 2*x + 3,
     type= "l"
     )
```
